```python
import pandas as pd

df = pd.DataFrame()
```

```python
# df.to
```

### to_csv

```python
df = pd.read_csv("../../2.Pandas/dataset/deliveries.csv")
df.head()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()

temp_df
```

```python
temp_df.to_csv("batsman_runs.csv", index=False)
```

```python
df.pivot_table(index='batsman', columns="bowling_team", values='batsman_runs', aggfunc='sum').to_csv("batsman_vs_bowling_team_runs.csv", index=True)
```

### to_excel

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()

temp_df
```

```python
temp_df.to_excel("batsman_runs.xlsx", index=False)
```

```python
# temp_df.to_excel("output.xlsx", sheet_name='batsman_runs') 
```

```python
temp_df2 = df.pivot_table(index='batsman', columns="bowling_team", values='batsman_runs', aggfunc='sum').to_csv("batsman_vs_bowling_team_runs.csv", index=True)
```

```python
with pd.ExcelWriter("output.xlsx") as writer:
   
   temp_df.to_excel(writer, sheet_name='batsmans_runs', index=False)
   temp_df2.to_excel(writer, sheet_name='batsman_vs_bowling_team_runs', index=True)
```

### to_html

```python
df.query('batsman_runs == 6').pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='count').to_html("batsman_vs_bowling_team_sixes.html", index=True)
```
