#### plotting multiple plots

```python

plt.plot(footballer['Year'],footballer['Ronaldo Goals'])
plt.plot(footballer['Year'],footballer['Messi Goals'])
```

```python
# labels title
plt.plot(footballer['Year'],footballer['Ronaldo Goals'])
plt.plot(footballer['Year'],footballer['Messi Goals'])

plt.title('Messi  Vs Ronaldo Career Comparison')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(footballer['Year'],footballer['Ronaldo Goals'],color='#D9F10F')
plt.plot(footballer['Year'],footballer['Messi Goals'],color='black')
# plt.plot(footballer['Year'],footballer['Messi Goals'],color='#FC00D6')

plt.title('Messi  Vs Ronaldo Career Comparison')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
```

```python
plt.plot(footballer['Year'],footballer['Ronaldo Goals'],color='#D9F10F',linestyle='solid')
plt.plot(footballer['Year'],footballer['Messi Goals'],color='#FC00D6',linestyle='dotted')
# plt.plot(footballer['Year'],footballer['Messi Goals'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Messi  Vs Ronaldo Career Comparison')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
```

```python
plt.plot(footballer['Year'],footballer['Ronaldo Goals'],color='#D9F10F',linestyle='solid',linewidth=4)
plt.plot(footballer['Year'],footballer['Messi Goals'],color='#FC00D6',linestyle='dotted',linewidth=2)
# plt.plot(footballer['Year'],footballer['Messi Goals'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Messi  Vs Ronaldo Career Comparison')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
```

```python
plt.plot(footballer['Year'],footballer['Ronaldo Goals'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D')
plt.plot(footballer['Year'],footballer['Messi Goals'],color='black',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Messi  Vs Ronaldo Career Comparison')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
```

```python
# legend -> location
plt.plot(footballer['Year'],footballer['Ronaldo Goals'],color="#E09886",linestyle='solid',linewidth=3,marker='D',markersize=8,label='Ronaldo')
plt.plot(footballer['Year'],footballer['Messi Goals'],color='black' ,linestyle='dashdot',linewidth=2,marker='o',label='Messi')

plt.title('Messi  Vs Ronaldo Career Comparison')
plt.xlabel('Year')
plt.ylabel('Goals Scored')

plt.legend()
# plt.legend(loc='upper right')
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# bar chart using data
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('dataset/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
plt.bar(df['batsman'], df['2017'])

### graphs are overlapping, we can use width and position to avoid that
```

```python
np.array(df.shape[0]) ## gives us an array of numbers from 0 to number of rows in the dataframe
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'])
```

```python
## plotting graphs simultaneously
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='black')
plt.bar(np.arange(df.shape[0]), df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
data = [23,45,100,20,49]
plt.pie(data)
```

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('dataset/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
