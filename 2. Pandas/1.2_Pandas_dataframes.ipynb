### find all the final winners of each season

```python
## Oneliner

# ipl[ipl['MatchNumber'] == 'Final']
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']] ## fancy indexing
```

```python
# how many super over finishes have occured
# ipl[ipl['SuperOver'] == 'Y'].shape[0]

ipl[ipl['SuperOver'] == 'Y']
```

```python
print(ipl[ipl['SuperOver'] == 'Y'].shape)  ## returns (num_rows, num_columns)
print(ipl[ipl['SuperOver'] == 'Y'].shape[0]) ## returns num_rows
```

```python
# how many matches has csk won in kolkata
## & : bitwise and operator for boolean indexing 
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
### Toss winner and match winner in same match
ipl[ipl['TossWinner'] == ipl["WinningTeam"]]
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner'] == ipl["WinningTeam"]].shape[0]/ipl.shape[0] * 100)
```

```python
# movies with rating higher than 8 and votes>10000
movies[(movies['imdb_rating'] > 8.5) & (movies['imdb_votes'] > 10000)].shape[0] 
```
