### Feature Selection using LASSO

```python
from sklearn import datasets
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/npradaschnor/Pima-Indians-Diabetes-Dataset/master/diabetes.csv')

df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

cols = X_train.columns

scaler = StandardScaler() ##scale the data to have mean 0 and variance 1 with range of -1 to 1
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = pd.DataFrame(X_train_scaled,columns=cols)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=cols)
```

```python
X_train_scaled.head()
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1) ##0.01

lasso.fit(X_train_scaled, y_train)
```

```python
lasso.coef_
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = pd.Series(np.abs(lasso.coef_),index=cols)

x.sort_values(ascending=False).plot(kind='bar')
```

### SelectFromModel

```python
from sklearn import datasets
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/npradaschnor/Pima-Indians-Diabetes-Dataset/master/diabetes.csv')

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
```

```python
from sklearn.feature_selection import SelectFromModel

sfm = SelectFromModel(model, threshold='mean')
# sfm = SelectFromModel(model, threshold=0.15)
```

```python
sfm.fit(X_train, y_train)
```

```python
sfm.get_support(indices=True)
```

```python
sfm.feature_names_in_[sfm.get_support(indices=True)]
```

```python
X_train_trans = sfm.transform(X_train)

X_train_trans = pd.DataFrame(X_train_trans,columns=sfm.feature_names_in_[sfm.get_support(indices=True)])

X_train_trans
```

### Sklearn RFE

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load iris dataset
url = "https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv"
df = pd.read_csv(url)

# Separate features and target variable
X = df.drop("species", axis=1)
y = df["species"]

# Initialize RandomForestClassifier
model = RandomForestClassifier()

# Initialize RFE
rfe = RFE(estimator=model, n_features_to_select=1)

# Fit RFE
rfe.fit(X, y)

# Print the ranking
ranking = rfe.ranking_
print("Feature ranking:")

for i, feature in enumerate(X.columns):
    print(f"{feature}: {ranking[i]}")
```

### Mutual Information 

- Feature based technique





```python
import pandas as pd

data = {
    'A': ['a1', 'a2', 'a1', 'a1', 'a2', 'a1', 'a2', 'a2'],
    'B': ['b1', 'b2', 'b2', 'b1', 'b1', 'b2', 'b2', 'b1']
}

df = pd.DataFrame(data)
```

```python
marginal_prob = pd.crosstab(df['A'], df['B'], margins=True, normalize='all')
marginal_prob
```

```python
from sklearn.feature_selection import mutual_info_classif ## used for classification problems to compute mutual information between features and target variable
from sklearn.datasets import load_iris
import pandas as pd
## mutual_info_regression is used for regression problems to compute mutual information between features and target variable

# Load iris dataset
iris = load_iris()
X = iris['data']
y = iris['target']

# Compute mutual information
mi = mutual_info_classif(X, y)

# Print mutual information
for i, mi_value in enumerate(mi):
    print(f"Feature {i}: Mutual Information = {mi_value}")
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.datasets import load_iris

# Load iris dataset
iris = load_iris()
X = iris['data']
y = iris['target']

# Create SelectKBest feature selector
selector = SelectKBest(mutual_info_classif, k=2)

# Fit and transform
X_new = selector.fit_transform(X, y)

# Get columns to keep and create new dataframe with those only
cols = selector.get_support(indices=True)

print(iris.feature_names)
print(cols)
```
