## np.flip

- This reverses the order of array elements along the specified axis, preserving the shape of the array

```python
np.flip(a)
```

```python
np.flip(b)
```

## np.put()

- Replaces specific elements of an array with given values of p_array. Array indexed works on flattened array

```python
np.put(a, [0,1], [100, 530]) ## permanently changes the items in the specified position
```

```python
a
```

## np.delete

- Returns a new array with the deletion of sub-arrays along with the mentioned axis.

```python
np.delete(a, 0) ## index
```

```python
np.delete(a,[0, 2, 4])
```

## set functions

- np.union1d
- np.intersect1d
- np.setdiff1d
- np.setxor1d
- np.in1d

```python
m = np.array([1,2,3,4, 5])
n = np.array([6,7,3,4, 5])
```

```python
np.union1d(m,n)
```

```python
np.intersect1d(m,n)
```

```python
np.setdiff1d(m,n)
```

```python
np.setxor1d(m,n)
```

```python
np.in1d(m, 1)
```

## np.clip

- function is used to clip (limit) the values in an array

```python
import numpy as np

a = np.random.randint(1, 100, 15)
a
```

```python
b = np.random.randint(1, 100, 24).reshape(6,4)
b
```

```python
np.clip(a, a_min=25, a_max=75)
```
