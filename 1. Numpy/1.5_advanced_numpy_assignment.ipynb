### `Q-1:` Find the nearest element in the array to a given integer.
```

```

### `Q-4:` Your are given an array which is havig some nan value. You job is to fill those nan values with most common element in the array.
```
arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])

```

````python
# code here

arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])
arr
````

````python
arr.flatten() ## converts to 1D array
````

````python

d = {}

## count the frequency of each element in the array
for i in arr.flatten():
    if  i in d:
        d[i] += 1
    
    else:
        d[i] = 1
        
arr[np.isnan(arr)] = sorted(d.items(), key=lambda x: x[1])[-1][0]

arr
        
````

### `Q-9:` Write a program implement Boradcasting Rule to check if two array can be added or not.
Given tuples of shapes.
```
shape of a- (3,2,2)
shape of b- (2,2)

check_broadcast(a, b) -> return Boolean (True if can broadcasted, False other wise.)
```

````python
# code here

## 3 4 2 1
##     1 4

## align both shapes


## loop for both shapes from behind

    ## if two numbers are not equal and none of the numbers are 1, return False

## return True

def check_broadcastable(a, b):
    
    print(a, b)
    a = a[::-1]
    # print(a)
    b = b[::-1]
    print(a, b)
    
    ## why the zip is done? -- to loop through both shapes simultaneously
    for ai, bi in zip(a,b):
        
        if ai != bi and ai != 1 and bi != 1:
            return False
        
    return True

check_broadcastable((4,1,3), (5,3))
````

````python
def check_broadcastable(shape1, shape2):
    len1 = len(shape1)
    len2 = len(shape2)
    
    for i in range(1, max(len1, len2)+1):
        dim1 = shape1[-i] if i <= len1 else 1
        dim2 = shape2[-i] if i <= len2 else 1
        
        if dim1 != dim2 and dim1 != 1 and dim2 != 1:
            return False
    
    return True
````
